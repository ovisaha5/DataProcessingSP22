####Navigating Data

```python
%%bigquery --project smart-tracer-338719
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 7
```

####Average Time for a Trip by Starting Point

```python
%%bigquery --project smart-tracer-338719
SELECT 
  AVG(duration_minutes) as avg_duration_minutes, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
```

####Total Number of Trip Based on Starting Point

```python
%%bigquery --project smart-tracer-338719
SELECT
COUNT(start_station_name) as total_trip_start, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
```
