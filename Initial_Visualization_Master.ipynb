####Loading data

```python
import pandas as pa
import seaborn as sns
import matplotlib.pyplot as plt
df= pa.read_csv('https://raw.githubusercontent.com/ovisaha5/DataProcessingSP22/main/covid-vaccinations-vs-gdp-per-capita.csv')
df.head(10)
```

####Initial Visualization of the Master

```python
df1=df.groupby('Entity') #Grouping the dataframe by countries
df2=df1.first()
df2
```

```python
sns.scatterplot(data = df2, x = "GDP per capita, PPP (constant 2011 international $)", y = "total_vaccinations_per_hundred", hue = "Continent", size= "Entity", sizes=(200,200))
plt.legend(bbox_to_anchor=(1.3, 1)) 
plt.ylim(-2, 100)
#plt.xlim(1000, 100000)
sns.set(rc = {'figure.figsize':(20, 15)})
plt.show()
```

Here on X axis I have selected the GDP per capita and Y axis I have selected total vaccination per hundred. Grouped the dataframe by countries.
