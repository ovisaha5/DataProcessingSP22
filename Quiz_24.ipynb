```python
import pandas as pa
import matplotlib
import numpy as np
df= pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')
df.head()
```

####Bar graph of the maximum 'price' by 'neighbourhood_group'

```python
ax=df.groupby('neighbourhood_group').price.agg('max').plot(kind='bar')

for container in ax.containers:
    ax.bar_label(container)
```

####Multiple bar graph with 'neighbourhood_group' and 'room_type' by looking at the average 'price'

```python
ax=df.groupby(['neighbourhood_group','room_type']).price.agg('mean').plot.bar(figsize=(12,8))
for container in ax.containers:
    ax.bar_label(container)
```

```python
df_pivot = pa.pivot_table(
    df, 
    values="price",
    index="room_type",
    columns="neighbourhood_group", 
    aggfunc=np.mean
)

ax=df_pivot.plot.bar(figsize=(17, 10))
for container in ax.containers:
    ax.bar_label(container)
```
