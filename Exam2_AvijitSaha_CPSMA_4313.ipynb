###Question 1

**a.** Storing data and comment on the appropriateness of each

```python
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head(10).T #Store the data in dataframe.
```

```python
df.dtypes #Examine the data types.
```

According to the data types provided in this dataframe I think, the time column need to be float instead of integer. total_steps can be changed as integer instead of float. min_heart_rate need to be changed as a float type. Also, the last five column vo2_max(%), aerobic(%), anarobic(%), intensive(%) and light(%), I will change it as a float instead of integer and the first column(activity_day) need to be DateTime data type.

**b.** Remove the 'Unnamed' column

```python
df = df.drop(columns="Unnamed: 0") #removing the 'Unnamed column'
df.head(10).T #using this command to see the column clearly
```

**c.** Clean the column names to remove the unit declaration, (%), using regular expressions. 

```python
df= df.rename(columns=lambda x: re.sub(r'[(.%)]','',x)) #cleaning column using regular expression
df.columns #lambda function help to pass the command for every column
```

```python
df.head().T
```

**d.** Converting the activity_day to DayTime format

```python
df.activity_day= pa.to_datetime(df.activity_day, format='%Y-%m-%d') #converting the activity_day column from an object to a datetime format
df.dtypes
```

**e.** Impute ‘total steps’ by replacing the ‘NaN’s with an appropriate number of steps and converting appropriate data types.

```python
df.total_steps=df.total_steps.fillna(method='pad')#filling it by the last value
```

```python
df.total_steps=df.total_steps.fillna(method='bfill') #filling it by the first value
df.total_steps.isnull().sum() #Calculating the total null value
```

```python
df.head(5).T #Showing the dataframe again 
```

```python
df.total_steps.astype('int') #Convert the total_step column from float to integer
```

**f.** Convert non-empty avg_pace to float

```python
df.avg_pace= pa.to_datetime(df.avg_pace)
df1= df.avg_pace.apply(lambda avg_pace: avg_pace.hour + (avg_pace.minute / 60)) #Converting the data from second to decimal
df1
```

```python
df1.dtypes 
```

It is now a float data type.

**g.** Group data by ‘workout type’ and find the mean, median, count and standard deviation of ‘calories’.

```python
df.groupby('workout_type').calories.agg(['mean','median','count', 'std']) #groupby work_out types for calories
```

**h.** Create an indicator column that identifies if the activity achieved 30% or more ‘aerobic’ activity.

```python
indicator = df.aerobic >=30   #Creating the dummy fuction for the indicator
df1=pa.concat([df,pa.DataFrame({'Indicator':indicator})], axis = 1)
df1.head(8).T
```

**i.** Max workout_type by day and work_out type

```python
df.groupby([df.activity_day.dt.day_name(),df.workout_type]).max_cadence.agg('max') #Using groupby to group it by date and workout_type
```
