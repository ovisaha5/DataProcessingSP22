```python
import requests
import pandas as pa
from bs4 import BeautifulSoup


r = requests.get('https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films')
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
```

####MCU Infinity Series Movie List

```python
tables = html_soup.find_all('table',class_="wikitable plainrowheaders")
df = pa.read_html(str(tables))[0]
df.columns=["Film","U.S. release date","Director(s)", "Screenwriter(s)","Producer(s)", "Unnamed: 5_level_0"]
df = df.drop(columns="Unnamed: 5_level_0")
df = df.drop([6,13])
df
```
