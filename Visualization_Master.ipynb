Renaming The coulmn might help me to add the country name on each dot.

```python
ax=sns.scatterplot(data = df2, x = "GDP per capita, PPP (constant 2011 international $)", y = "total_vaccinations_per_hundred", hue = "Continent", s=200)
plt.legend(bbox_to_anchor=(1.15, 1)) 
plt.ylim(-2, 100)
sns.set(rc = {'figure.figsize':(20, 15)})
ax.set_xscale('log') #Here I have set the x-axis logarithmic which change the entire scale of the graph
ax.set_yscale('linear')

def label_point(x, y, val, ax):#fuction to add the country name.
    a = pa.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.02, point['y'], str(point['val']))
label_point(df3.GDP_Per_Capita, df3.total_vaccinations_per_hundred, df3.Entity, plt.gca())
plt.show()
```
