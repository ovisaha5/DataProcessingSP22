```python
import pandas as pa
df= pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')
```

```python
df.price.plot.hist(title="Histogram of the Price", bins=250, xlim=[0, 1200]).set_xlabel("Price") 
#set bin to 250 and limt the x axix from0 to 1200 to get more accurate data
```

```python
df.groupby('neighbourhood_group').price.plot.hist(title="Price By Neighbourhood Group", bins=250, xlim=[0, 1200], xlabel="Price", legend="True")
#Here 250 bin give the data clearly
```
