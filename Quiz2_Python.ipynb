#### Importing libraries

```python
import pandas as pd
```

#### Loading Irish Dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/iris.csv')

df
```

#### Calling A Column

```python
df[['SepalLength']]
```

Find Mean

```python
df.SepalLength.mean()
```

#### Last Row

```python
df.iloc[-1]
```

#### Melt the Data

```python
df_melt = df.melt(id_vars = 'Class',
                  value_vars = ['SepalLength','PedalLength'],
                  var_name = 'colmuns')
df_melt.tail()
```

```python
df_melt.pivot(columns = 'colmuns', values = ['value', 'Class'])
```

####Select Flower

```python
df.query('PedalLength >5 and SepalLength <7')
```

#### Group The Data

```python
df.groupby(by = 'Class').agg(['mean','median'])
```
