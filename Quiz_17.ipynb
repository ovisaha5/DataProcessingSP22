```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/iris.csv')
df
```

#####Identifying the greater mean with float or integer

```python
df.SepalLength.mean()
```

```python
df.SepalLength=df.SepalLength.astype('int')
df
```

```python
df.SepalLength.mean()
```

So, the integer decrease the mean.

#####Direction of the shift with GROUPBY Class

####Gathering the mean, median, count and standard deviation of all column

```python
df.groupby('Class').agg(['mean','median','count', 'std'])
```
