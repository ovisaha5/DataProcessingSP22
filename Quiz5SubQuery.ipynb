#####Second Most Popular

```python
%%bigquery --project smart-tracer-338719
SELECT
COUNT(*) as total_trip_start, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY total_trip_start DESC
```

```python
%%bigquery --project smart-tracer-338719
SELECT
 MAX(total_trip_start) as max_trip
 FROM (SELECT
      COUNT(start_station_name) as total_trip_start, start_station_name
      FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
      GROUP BY start_station_name)
```

```python
%%bigquery --project smart-tracer-338719
SELECT MAX (total_trip_start) as second_most_popular
FROM (SELECT
      COUNT(*) as total_trip_start, start_station_name
      FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
      GROUP BY start_station_name
      ORDER BY total_trip_start DESC) 
WHERE total_trip_start != (SELECT
      MAX(total_trip_start) as max_trip
      FROM (SELECT
      COUNT(start_station_name) as total_trip_start, start_station_name
      FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
      GROUP BY start_station_name))
```

####Trip Lasted Over an Hour

```python
%%bigquery --project smart-tracer-338719
SELECT start_station_name, duration_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name AND duration_minutes = 60
```
