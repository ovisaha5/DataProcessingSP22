```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head()
```

###Scatter Plot

```python
ax = sns.scatterplot(data = df, x = 'distance', y = 'calories') 
ax.set(title = "Time vs Calorie",
       )
plt.show()
```

```python
sns.scatterplot(data = df, x = 'distance', y = 'calories', hue = "workout_type", size="distance") #multiple variable more than two
plt.legend(bbox_to_anchor=(1.05, 1))
```

```python
sns.scatterplot(data = df, x = 'distance', y = 'calories', hue = "workout_type", size="time", style='workout_type')
plt.legend(bbox_to_anchor=(1.05, 1)) #this is an example of 5 variable in one graph
```

```python
sns.lmplot(data = df, 
                x = 'time', 
                y = 'calories',
                hue = 'workout_type',
                ci = False )
```

Yes, it is possible to get 5 or 6 variable in one graph by scatter plot.
