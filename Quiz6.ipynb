#####Joining by Inner Join

```python
%%bigquery --project smart-tracer-338719
SELECT start_station_name, property_type, duration_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations` as stations
JOIN `bigquery-public-data.austin_bikeshare.bikeshare_trips` as trips ON name=start_station_name
GROUP BY station_id, start_station_name, property_type, duration_minutes
ORDER BY duration_minutes DESC
LIMIT 50
```

####Average time

```python
%%bigquery --project smart-tracer-338719
SELECT 
AVG(duration_minutes) as avg_duration_mins, property_type
FROM(SELECT start_station_name, property_type, duration_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations` as stations
JOIN `bigquery-public-data.austin_bikeshare.bikeshare_trips` as trips ON name=start_station_name
GROUP BY station_id, start_station_name, property_type, duration_minutes
)
GROUP BY property_type
ORDER BY avg_duration_mins DESC
```
